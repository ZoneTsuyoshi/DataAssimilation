# 粒子スムーザーの実装

```python
class ParticleFilter(object):
    def __init__(self, y, n_particle = 1000, sigma_2 = 0.01, alpha_2 = 0.01, mu_0 = 0.0):
        self.y = y
        self.n_particle = n_particle
        self.sigma_2 = sigma_2
        self.alpha_2 = alpha_2
        self.mu_0 = mu_0
        self.log_likelihood = - np.inf
    
    # likelihood for normal distribution
    def norm_likelihood(self, y, x, s2):
        return (np.sqrt(2 * np.pi * s2))**(-1) * np.exp(-(y - x)**2 / (2 * s2))

    def F_inv(self, w_cumsum, idx, u):
            if np.any(w_cumsum < u) == False:
                return 0
            k = np.max(idx[w_cumsum < u])
            return k + 1
        
    def resampling(self, weights):
        w_cumsum = np.cumsum(weights)
        idx = np.asanyarray(range(self.n_particle)) # labelの生成
        k_list = np.zeros(self.n_particle, dtype = np.int32) # サンプリングしたkのリスト格納場所
        
        # 一様分布から重みに応じてリサンプリングする添え字を取得
        for i, u in enumerate(rd.uniform(0, 1, size = self.n_particle)):
            k = self.F_inv(w_cumsum, idx, u)
            k_list[i] = k
        return k_list

    def resampling2(self, weights):
        """
        計算量の少ない層化サンプリング
        """
        idx = np.asanyarray(range(self.n_particle))
        u0 = rd.uniform(0, 1 / self.n_particle)
        u = [1 / self.n_particle*i + u0 for i in range(self.n_particle)]
        w_cumsum = np.cumsum(weights)
        k = np.asanyarray([self.F_inv(w_cumsum, idx, val) for val in u])
        return k
    
    def simulate(self, y, seed = 71):
        '''
        y : observation data
        T : length of y
        x : prediction distribution particles
        x_resampled : filter distribution particles
        initial_x : initial distribution of x
        w : weight lambda of each particle
        w_normed : normed weitht beta of each particle
        l : log likelihood for each time
        v : system noise particles
        k : index number for resampling
        '''
        rd.seed(seed)

        # 時系列データ数
        # if y is 1 dimension data
        T = len(y)
        
        # 潜在変数
        x = np.zeros((T + 1, self.n_particle))
        x_resampled = np.zeros((T + 1, self.n_particle))
        
        # 潜在変数の初期値
        initial_x = rd.normal(self.mu_0, 1, size = self.n_particle)
        x_resampled[0] = initial_x
        x[0] = initial_x

        # 重み
        w        = np.zeros((T, self.n_particle))
        w_normed = np.zeros((T, self.n_particle))

        l = np.zeros(T) # 時刻毎の尤度

        for t in range(T):
            print("\r calculating... t={}".format(t + 1) + "/" + str(T), end="")
            for i in range(self.n_particle):
                # 1階差分トレンドを適用
                v = rd.normal(0, np.sqrt(self.alpha_2 * self.sigma_2)) # System Noise
                x[t + 1, i] = x_resampled[t, i] + v # システムノイズの付加
                w[t, i] = self.norm_likelihood(y[t], x[t + 1, i], self.sigma_2)
                # y[t]に対する各粒子の尤度
            w_normed[t] = w[t] / np.sum(w[t]) # 規格化
            l[t] = np.log(np.sum(w[t])) # 各時刻対数尤度

            # Resampling
            #k = self.resampling(w_normed[t]) # リサンプルで取得した粒子の添字
            k = self.resampling2(w_normed[t]) # リサンプルで取得した粒子の添字（層化サンプリング）
            x_resampled[t + 1] = x[t + 1, k]
            
        # 全体の対数尤度
        self.log_likelihood = np.sum(l) - T * np.log(n_particle)
        
        #self.x = x
        #self.x_resampled = x_resampled
        #self.w = w
        #self.w_normed = w_normed
        #self.l = l
        return x, x_resampled, w, w_normed, l
        
    def get_filtered_value(self, y, x, w_normed):
        """
        尤度の重みで加重平均した値でフィルタリングされ値を算出
        """
        #x, _, _, w_normed, _ = self.simulate(y)
        return np.diag(np.dot(w_normed, x[1:].T))
        
    def draw_graph(self):
        # グラフ描画
        T = len(self.y)
        x, _, _, w_normed, l = self.simulate(self.y)
        
        plt.figure(figsize=(16,8))
        plt.plot(range(T), self.y, label = "Y")
        plt.plot(self.get_filtered_value(self.y, x, w_normed), "g", label = "PF")
        
        for t in range(T):
            plt.scatter(np.ones(self.n_particle)*t, x[t], color = "r", s = 2,
                        alpha = 0.1)
        
        plt.legend(loc = 'upper right')
        plt.xlabel('time')
        plt.ylabel('value')
        plt.title("sigma^2={0}, alpha^2={1}, log likelihood={2:.3f}"
                  .format(self.sigma_2, self.alpha_2, self.log_likelihood))
```

```python
class ParticleSmoother(ParticleFilter):
    '''
    y : observation data
    n_particle : number of particles which use PF / PS
    sigma_2 : variance of observation noise
    alpha_2 : relatively variance = (variance of system noise) / (variance of observation noise)
    mu_0 : initial x_0 mean
    lag : smoother lag
    '''
    def __init__(self, y, n_particle = 1000, sigma_2 = 0.01, alpha_2 = 0.01, mu_0 = 0.0,
                 lag = 20):
        ParticleFilter.__init__(self, y, n_particle, sigma_2, alpha_2, mu_0)
        self.lag = lag
    
    def particle_smoother(self, y) :
        '''
        T : length of y
        before_y : y before fixed lag smoothing
        center_y : not before_y, after_y
        after_y : y after fixed lag smoothing
        before_x : x before fixed lag smoothing
        before_x_resampled : filter x before fixed lag smoothing
        before_w : weight lambda before fiexed lag smoothing
        before_w_normed : normed weight beta before fixed lag smoothing
        before_l : log likehood before fixed lag smoothing
        tilde_x : expansion state vector for prediction
        tilde_x_resampled : expnasion state vector for filtering
        smooth_x : smoothing x_{t-L|t}
        '''
        T = len(y)
        #before__y = y[:self.lag]
        #center_y = y[self.lag:T - self.lag]
        #after_y = y[T - self.lag:]
        
        # prediction and filtering before lag smoothing
        initial_x = rd.normal(self.mu_0, 1, size = self.n_particle)
        before_x, before_x_resampled, _, _, _ = self.simulate(y[:self.lag])
        #x, x_resampled, w, w_normed, l = self.simulate(y)
        tilde_x = np.zeros((T + 1 - lag, self.n_particle, lag + 1))
        tilde_x_resampled = np.zeros((T + 1 - lag, self.n_particle, lag + 1))
        # tilde_x means tilde x_lag
        tilde_x[0] = before_x.T
        tilde_x_resampled[0] = before_x_resampled.T
        
        # smoothing x
        smooth_x = np.zeros((T + 1, self.n_particle))
        
        # weight
        w        = np.zeros((T, self.n_particle))
        w_normed = np.zeros((T, self.n_particle))
        
        for t in range(T - lag):
            print("\r calculating... smoothing t={}".format(t + 1) + "/" + str(T - lag), end="")
            for i in range(self.n_particle):
                # 1階差分トレンドを適用
                v = rd.normal(0, np.sqrt(self.alpha_2 * self.sigma_2)) # System Noise
                tilde_x[t + 1, i, lag] = tilde_x_resampled[t, i, lag] + v # システムノイズの付加
                tilde_x[t + 1, i, :lag] = tilde_x_resampled[t, i, :lag]
                w[t, i] = self.norm_likelihood(y[t + lag], tilde_x[t + 1, i, lag], self.sigma_2)
                # y[t]に対する各粒子の尤度
            w_normed[t] = w[t] / np.sum(w[t]) # 規格化
            #l[t] = np.log(np.sum(w[t])) # 各時刻対数尤度

            # Resampling
            #k = self.resampling(w_normed[t]) # リサンプルで取得した粒子の添字
            k = self.resampling2(w_normed[t]) # リサンプルで取得した粒子の添字（層化サンプリング）
            tilde_x_resampled[t + 1] = tilde_x[t + 1, k]
        
        smooth_x[:T - lag + 1, :] = tilde_x[:, :, 0]
        # after lag smoothing, normal prediction
        smooth_x[T - lag + 1:, :] = tilde_x[T - 2*lag + 1:, :, lag]
        w_normed[T - lag:, :] = w_normed[T - 2*lag:T - lag, :]
        
        return smooth_x, w_normed
    
    def draw_graph_smoothing(self, x, w_normed, smooth_x, smooth_w_normed) :
        # グラフ描画
        T = len(self.y)
        #x, _, _, w_normed, l = self.simulate(self.y)
        #smooth_x, smooth_w_normed = self.particle_smoother(self.y)
        
        plt.figure(figsize=(16,8))
        plt.plot(range(T), self.y, "b", label = "Y")
        plt.plot(self.get_filtered_value(self.y, x, w_normed), "g", label = "PF")
        plt.plot(self.get_filtered_value(self.y, smooth_x, smooth_w_normed), "m", label = "PS")
        
        for t in range(T):
            plt.scatter(np.ones(self.n_particle)*t, smooth_x[t], color = "r", s = 2,
                        alpha = 0.1)
        
        plt.legend(loc = 'best')
        plt.xlabel('time')
        plt.ylabel('value')
        plt.title("sigma^2={0}, alpha^2={1}, log likelihood={2:.3f}"
                  .format(self.sigma_2, self.alpha_2, self.log_likelihood))
```

```python
# ハイパーパラメーター
a = -2
b = -1
n_particle = 10**3 * 5
sigma_2 = 2**a
alpha_2 = 10**b
mu_0 = 15
lag = 20
```

```python
ps = ParticleSmoother(df.data.values, n_particle, sigma_2, alpha_2, mu_0, lag)
x, _, _, w_normed, l = ps.simulate(ps.y)
smooth_x, smooth_w_normed = ps.particle_smoother(ps.y)
```

```python
ps.draw_graph_smoothing(x, w_normed, smooth_x, smooth_w_normed)
plt.savefig("plot/171022_particle_smoother_yields.png")
plt.show()
```

## Kalman Smootherとの比較

```python
from pykalman import KalmanFilter
kf = KalmanFilter(transition_matrices = np.array([[1, 1], [0, 1]]),
                  transition_covariance = 0.0000001*np.eye(2))

kalman_smooth_x = kf.em(df.data.values).smooth(df.data.values)[0]
kalman_filter_x = kf.em(df.data.values).filter(df.data.values)[0]

plt.figure(figsize=(16, 8))
plt.plot(df.data.values, "-xc", label="observations")
plt.plot(kalman_smooth_x[:, 0], "b", label="smoothed")
plt.plot(kalman_filter_x[:, 0], "r", label="filtered")
#plt.axhline(x, color="k", label="truth value")
plt.legend(loc = 'best')
plt.title('Kalman Filter and Smoothing')
plt.savefig('plot/171022_kalman_smoothing_yields.png')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.plot(df.data.values, "-xc", label="observations")
plt.plot(kalman_smooth_x[:, 0], "b", label="KS")
plt.plot(ps.get_filtered_value(ps.y, smooth_x, smooth_w_normed), "r", label = "PS")
plt.legend(loc = 'best')
plt.title('Kalman Smoothing and Particle Smoothing')
plt.savefig('plot/171022_particle_smoothing_yields_with_KS.png')
plt.show()
```

Kalman Smoothing と恐ろしく結果が一致していない．<br>
最後のラグの分は良いとしても，その前までの低空飛行はどうなってるんだろう．考えてみた．<br>
Kalman Smoothing では固定区間平滑化：すなわち全てのデータを使っている．<br>
一方，Particle Smoothing では拡張状態ベクトルを定義して固定ラグ平滑化：ラグ分しかデータを支えていない．<br>
それにしても，最初の方は低空飛行し過ぎでは？と疑問が湧く．
